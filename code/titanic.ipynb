## Traiter la variable Age

###### Pour la variables age nous avons 177 valeur manquantes. Comme notre dataset trains n'es pas tres grand nous allons
###### remplacer les valeurs manquante par la mediane 

```python
train_df.loc[train_df['Age'].isna(), 'Age'] = train_df['Age'].median()
test_df.loc[test_df['Age'].isna(), 'Age'] = test_df['Age'].median()
```

##  Traiter la variable Cabin

###### Pour la variables cabin le problemes est qu'il y'a  parfois 2 ou 3 cabin  par ligne.
###### Nous allons prendre la premier cabin pour resoudre ce probleme .

```python
train_df['Cabin'] = train_df['Cabin'].str.split(',', n=1, expand=True)[0]
display(train_df)
```

##### La valeur A0 representes les NA de cabin


```python
train_df['Cabin'].fillna('A0', inplace=True)
```

#### pour les Na de Embarked  on remplace par S 

```python
train_df['Embarked'].fillna('S', inplace=True)
```
